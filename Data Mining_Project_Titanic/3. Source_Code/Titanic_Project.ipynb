### Mengisi Missing value pada categorical data 

#### Mengisi Value pada **Embarked**

```python
train[train['Embarked'].isnull()]
```

```python
train['Embarked'].value_counts()
```

```python
mode_embarked = train['Embarked'].mode()[0] #Rumus Mencari Modus
mode_embarked
```

```python
train['Embarked'] = train['Embarked'].fillna(mode_embarked)
```

```python
train[train['Embarked'].isnull()]
```

```python
train[train['Embarked'].index == 829]
```

### Mengisi Missing value pada Numerik data 

#### Mengisi Value pada **Umur**

```python
mean_age=train['Age'].mean() #Rumus Mencari rata-rata
mean_age
```

```python
train['Age'] = train['Age'].fillna(mean_age)
```

```python
mean_age_test=test['Age'].mean() #Rumus Mencari rata-rata
mean_age_test
```

```python
test['Age'] = test['Age'].fillna(mean_age_test)
```

#### Mengisi Value pada **Fare**

```python
mean_fare_test = test['Fare'].mean()

test['Fare'] = test['Fare'].fillna(mean_fare_test)
```

#### Drop Data Cabin Karena terlalu banyak Missing Value

```python
train = train.drop(['Cabin'], axis = 1)
test = test.drop(['Cabin'], axis = 1 )
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

## **Convert Data to Numeric**

### **Convert Data Sex**

```python
train.head()
```

```python
sex_mapping = {"male" : 0, "female" : 1}

train['Sex'] = train['Sex'].map(sex_mapping)
test['Sex'] = test['Sex'].map(sex_mapping)
```

```python
train.head()
```

### **Convert Data Embarked**

```python
embarked = train[['Embarked']] 

dummies = pd.get_dummies(embarked)
```

```python
dummies.head()
```

```python
train.head()
```

```python
train = pd.merge(train.reset_index(),  dummies.reset_index())
train.head()
```

```python
train = train.drop(["index",'Embarked'], axis =1)
```

Lakukan yang sama pada data test

```python
embarked_test = test[['Embarked']] 

dummies_test = pd.get_dummies(embarked_test)
```

```python
test = pd.merge(test.reset_index(),  dummies_test.reset_index())
test = test.drop(['index', 'Embarked'], axis = 1 )
test.head()
```

### **Extract Title from Name**

```python
train['Title'] = train['Name'].str.extract(' ([A-Za-z]+)\.', expand = False)
```

```python
train.head()
```

```python
train['Title'].unique()
```

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }

train['Title'] = train['Title'].map(title_mapping)
```

```python
train.head()
```

Lakukan juga pada data test

```python
test['Title'] = test['Name'].str.extract(' ([A-Za-z]+)\.', expand = False)
test['Title'] = test['Title'].map(title_mapping)
```

```python
  # Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
train.shape, test.shape
```

```python
train_data = train.drop("Survived",axis = 1 )
target = train['Survived']
```

##**Naive Bayes Klasifikasi**

```python
k_fold = KFold(n_splits=5 , shuffle=True, random_state=0)

clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

##**Decision Tree Clasification**

```python
k_fold = KFold(n_splits=5 , shuffle=True, random_state=0)

clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```
